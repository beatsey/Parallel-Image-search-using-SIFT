```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
from mpi4py import MPI

%matplotlib inline
```

```python
%%writefile test.py

import matplotlib.pyplot as plt
import numpy as np
from mpi4py import MPI


def shift_pic_parallel(img, shiftBy):
    comm = MPI.COMM_WORLD
    size = comm.Get_size()
    rank = comm.Get_rank()
    
    shape = img.shape
    
    if rank == 0:
        start = MPI.Wtime()

    img = img[:,shape[1] * rank // size:shape[1] * (rank + 1) // size]
    img = np.append(img[shiftBy:],img[:shiftBy],axis=0)
    
    #np.transpose(np.append(np.transpose(img,(1,0,2)),np.transpose(im[1],(1,0,2))).reshape(img.shape[1],img.shape[0],img.shape[2]),(1,0,2))
    
    img = comm.reduce(np.transpose(img,(1,0,2)), op=np.append, root=0)

    if rank == 0:
        img = np.transpose(np.asarray(img).reshape((shape[1],shape[0],shape[2])),(1,0,2))
        end = MPI.Wtime()
        print(f'Execution time ({size} processes):', end - start)
        plt.imshow(img)
        plt.plot()
        print('Saving shifted image..')
        plt.imsave('shifted_img.png', img)
        return img

img = plt.imread('catWallpaper.jpg')
shift_pic_parallel(img,150)
```
